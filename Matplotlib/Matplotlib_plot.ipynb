1. Matplotlib:

a. Thư viện matplotlib.pyplot dùng để vẽ một đường thẳng từ tọa độ điểm X đến tọa độ điểm Y.
Ví dụ vẽ từ điểm có tọa độ A[2,4] đến điểm B[105, 120] ta phải import vào 2 mảng X([2, 105]) và Y([4, 120])

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([2, 105])
Y = np.array([4, 120])

plt.plot(X, Y)
plt.show
```

* Ví dụ 2: Vẽ các đường thẳng nối các điểm (1, 3) đến (2, 8) sau đấy đến (6, 1) và đến (8, 10):

```python
X = np.array([1,2,6,8])
Y = np.array([3,8,1,10])
plt.plot(X, Y)
plt.show()
```

Nếu không chỉ định cột X thì mặc định giá trị sẽ là 0,1,2,3

```python
plt.plot(Y)
plt.show()
```
