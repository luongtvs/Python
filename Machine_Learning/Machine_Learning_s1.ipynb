```python
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score

# Mount to google drive
drive.mount('/content/drive')

# Open file csv in google drive
csv_data = pd.read_csv('/content/drive/MyDrive/Python/Data/budget_movie.csv')

xproduction_budget =  np.array(csv_data['production_budget_usd'])
yworldwide_gross = list(csv_data['worldwide_gross_usd'])


slope, intercept, r, p, std_err = stats.linregress(xproduction_budget, yworldwide_gross)
def myfunct(production_budget):
    return slope*production_budget + intercept

prmodel = list(map(myfunct, xproduction_budget))

print('---Linear Regression ----')
print('r = ',r)
plt.scatter(xproduction_budget, yworldwide_gross, color = '#88c999')
plt.plot(xproduction_budget, prmodel, color = "red")
plt.ylabel('worldwide_gross')
plt.xlabel('production_budget')
plt.xlim(0, 500000000)
plt.ylim(0, 3000000000)



print('---Linear Polynomial Regression ----')
lnmodel = np.poly1d(np.polyfit(xproduction_budget, yworldwide_gross, 3))
print('r = ', r2_score(yworldwide_gross, lnmodel(xproduction_budget)))

predict_worldwide_gross = myfunct(15000000)
print(predict_worldwide_gross)
myline = np.linspace(0, 500000000, 1000)
plt.plot(myline, lnmodel(myline))
plt.show()
speed = lnmodel(7000000)
print(speed)

```
